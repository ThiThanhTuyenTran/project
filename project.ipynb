```python
import tkinter as tk
from tkinter import messagebox
```

```python
class TicTacToe:
    def __init__(self):
        self.current_player = 'X'
        self.board = [['', '', ''], ['', '', ''], ['', '', '']]
        self.root = tk.Tk()
        self.root.title('Tic-Tac-Toe')
        self.buttons = []
        for i in range(3):
            row = []
            for j in range(3):
                button = tk.Button(self.root, text='', width=10, height=5,
                                   command=lambda row=i, col=j: self.button_click(row, col))
                button.grid(row=i, column=j)
                row.append(button)
            self.buttons.append(row)
        self.reset_button = tk.Button(
            self.root, text='Reset', command=self.reset)
        self.reset_button.grid(row=3, column=1)
        self.label = tk.Label(
            self.root, text=f'Current player: {self.current_player}')
        self.label.grid(row=3, column=0)

    def reset(self):
        self.current_player = 'X'
        self.board = [['', '', ''], ['', '', ''], ['', '', '']]
        self.update_board()
        self.label.configure(text=f'Current player: {self.current_player}')

    def button_click(self, row, col):
        if self.board[row][col] == '':
            self.board[row][col] = self.current_player
            self.update_board()
            winner = self.check_winner()
            if winner:
                messagebox.showinfo('Winner', f'Player {winner} wins!')
                self.reset()
            elif self.check_tie():
                messagebox.showinfo('Tie', 'The game is a tie!')
                self.reset()
            else:
                self.switch_player()

    def update_board(self):
        for i in range(3):
            for j in range(3):
                self.buttons[i][j].configure(text=self.board[i][j])

    def switch_player(self):
        self.current_player = 'O' if self.current_player == 'X' else 'X'
        self.label.configure(text=f'Current player: {self.current_player}')

    def check_winner(self):
        for i in range(3):
            if self.board[i][0] == self.board[i][1] == self.board[i][2] != '':
                return self.board[i][0]
            if self.board[0][i] == self.board[1][i] == self.board[2][i] != '':
                return self.board[0][i]
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != '':
            return self.board[0][0]
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != '':
            return self.board[0][2]
        return None

    def check_tie(self):
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == '':
                    return False
        return True

    def play(self):
        self.root.mainloop()


if __name__ == '__main__':
    game = TicTacToe()
    game.play()
```
